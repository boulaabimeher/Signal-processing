```python
num_cols = data0.shape[1]  # Number of columns

fig = plt.figure(figsize=(20, 6*num_cols))
gs = GridSpec(num_cols, 2, width_ratios=[3, 1])

for i in range(num_cols):
    # Plotting entire data on the left
    ax1 = plt.subplot(gs[i, 0])
    col_name = data0.columns[i]

    ax1.plot(data0.index, data0[col_name], color='blue', label='Entire Data')
    ax1.fill_between(extracted_subset.index, data0[col_name].min(), data0[col_name].max(),
                     where=(extracted_subset.index >= extracted_subset.index.min()) &
                           (extracted_subset.index <= extracted_subset.index.max()),
                     alpha=0.7, color='red', label='Extracted Zone', zorder=2)
    ax1.legend()
    ax1.set_title(f'{col_name} - Scenario 01')

    # Plotting extracted data on the right
    ax2 = plt.subplot(gs[i, 1])
    ax2.plot(extracted_subset.index, extracted_subset[col_name], color='red', label='Extracted Subset')
    ax2.legend()
    ax2.set_title(f'{col_name} - Scenario 02')

plt.tight_layout()
plt.show()
```
